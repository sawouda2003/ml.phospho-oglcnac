```python
import requests
import json
import pandas as pd
import re
from functools import reduce
```

```python
search_url = "https://api.glygen.org/supersearch/search"
list_url = "https://api.glygen.org/supersearch/list"


payload = {
  "concept_query_list": [
    {
      "concept": "site",
      "query": {
        "aggregator": "$or",
        "aggregated_list": [],
        "unaggregated_list": [
          {
            "path": "phosphorylation_flag",
            "order": 1,
            "operator": "$eq",
            "string_value": "true"
          },
          {
            "path": "glycosylation_flag",
            "order": 0,
            "operator": "$eq",
            "string_value": "true"
          }
        ]
      }
    }
  ]
}

# the payload is the query for the Supersearch search

response = requests.post(url=search_url, json=payload)

print(response.text)

# printing out the json result
```

```python
search_data = response.json() #converting response to a json file
list_id = search_data["results_summary"]["site"]["list_id"] #obtaining the id


list_response = requests.post(url=list_url, json= {"id" : list_id})
print(list_response.text)

list_data = list_response.json()
```

```python
filtered_list = []

for item in list_data["results"] :
    if item["tax_id"] == 9606:
        print(item)
        filtered_list.append(item)

#adding only proteins with tax id 9606 to the list
```

```python
site_list = []

for item in filtered_list:
    if item["phosphorylation"] == "yes" or (item["glycosylation"] == "yes" and item["phosphorylation"] == "no") or item["glycosylation"] == "no" or (item["glycosylation"] == "yes" and item["phosphorylation"] == "yes") or (item["glycosylation"] == "yes" and item["glycosylation_type"] == "o-linked|o-glcnacylation"):
        print(item)
        site_list.append(item)
    else:
        print("no sites found")
```

```python
# For creating dataframe

# Defined the columns for the extracted information
columns = ["uniprotkb_canonical_ac", "protein_name", "site", "amino_acid", "phosphorylation", "glycosylation_type", "peptide_seq_five_before",
           "peptide_seq_five_after", "peptide_seq_ten_before", "peptide_seq_ten_after", "taxonomy_id"]

# Extracted the information from filtered_list_remove and store it in a list of dictionaries
extracted_data = []
for item in site_list:
    extracted_data.append({
        "uniprotkb_canonical_ac": item.get("uniprot_canonical_ac", ""),
        "protein_name": item.get("protein_name", ""),
        "site": item.get('start_pos', ''),
        "amino_acid": item.get("residue", ""),
        "phosphorylation": item.get("phosphorylation", ""),
        "glycosylation_type": item.get("glycosylation_type", ""),
        "peptide_seq_five_before": item.get("up_seq", "")[-5:],
        "peptide_seq_five_after": item.get("down_seq", "")[:5],
        "peptide_seq_ten_before": item.get("up_seq", "")[-10:],
        "peptide_seq_ten_after": item.get("down_seq", "")[:10],
        "taxonomy_id": item.get("tax_id", "")
    })

# Created a DataFrame from the extracted data
df = pd.DataFrame(extracted_data, columns=columns)

# Exported the DataFrame to an Excel file
df.to_excel("extracted_data.xlsx", index=False)
```

```python
# extracting potential o-glycosylation sites
pd.set_option('display.max_columns', 10000)
pd.set_option('display.max_rows', 10000)

df2 = df #making a copy of the original data frame
for i in df2['uniprotkb_canonical_ac']:
    response = requests.post(url= ("https://api.glygen.org/protein/detail/" + i ), json={"uniprot_canonical_ac": i }) #the post url require the id after the detail, extracting id from table
    protein_data = response.json()
    fasta_seq = protein_data["sequence"]['sequence']

    for g in protein_data['gene_names']: #retrieving gene names
        if (g['resource'] == 'UniProtKB') and (g['type'] == 'recommended'):
            gene_name = g['name']
            break
        else:
            if (g['resource'] == 'UniProtKB'):
                gene_name = g['name']
                break
    df2.loc[df2.uniprotkb_canonical_ac == i,["gene_name"]] = gene_name #adding gene names to the dataset

    kinase_list = []
    for p in protein_data['phosphorylation']:
        if 'kinase_gene_name' in p:
            kinase_list.append(p['kinase_gene_name']) #extracting kinase gene names
    kinase_gene_name = ';'.join(map(str, kinase_list))
    df2.loc[df2.uniprotkb_canonical_ac == i,["kinase_gene_name"]] = kinase_gene_name #adding kinase gene names to dataset

    # using re.finditer() to find all occurrences of T or S in fasta
    s_occurrences = re.finditer("S", fasta_seq)
    # using reduce() to get start indices of all occurrences
    s_res = reduce(lambda x, y: x + [y.start()], s_occurrences, [])
    t_occurrences = re.finditer("T", fasta_seq)
    # using reduce() to get start indices of all occurrences
    t_res = reduce(lambda x, y: x + [y.start()], t_occurrences, [])
    for j in s_res:
        if (j == df2.loc[df2.uniprotkb_canonical_ac == i, 'site'].values[0]):
            s_res.remove(j)
    for k in t_res:
        if k == df2.loc[df2.uniprotkb_canonical_ac == i, 'site'].values[0]:
            t_res.remove(k)

    #adding in rows for each new S site
    for s in s_res:
        s_dict = {
        "uniprotkb_canonical_ac": i,
        "protein_name": df2.loc[df2.uniprotkb_canonical_ac == i, 'protein_name'].values[0],
        "site": s, #s is the number position of the amino acid Ser
        "amino_acid": "Ser",
        "peptide_seq_five_before": fasta_seq[s-5:s],
        "peptide_seq_five_after": fasta_seq[s:s+5],
        "peptide_seq_ten_before": fasta_seq[s-10:s],
        "peptide_seq_ten_after": fasta_seq[s:s+10],
        "taxonomy_id": df2.loc[df2.uniprotkb_canonical_ac == i, 'taxonomy_id'].values[0],
        "gene_name": gene_name,
        "kinase_gene_name": kinase_gene_name
        }
        s_df = pd.DataFrame([s_dict])
        df2 = pd.concat([df2, s_df], ignore_index=True)

    #adding in rows for each new T site
    for t in t_res:
        t_dict = {
        "uniprotkb_canonical_ac": i,
        "protein_name": df2.loc[df2.uniprotkb_canonical_ac == i, 'protein_name'].values[0],
        "site": t, #t is the number position of the amino acid Thr
        "amino_acid": "Thr",
        "peptide_seq_five_before": fasta_seq[t-5:t],
        "peptide_seq_five_after": fasta_seq[t:t+5],
        "peptide_seq_ten_before": fasta_seq[t-10:t],
        "peptide_seq_ten_after": fasta_seq[t:t+10],
        "taxonomy_id": df2.loc[df2.uniprotkb_canonical_ac == i, 'taxonomy_id'].values[0],
        "gene_name": gene_name,
        "kinase_gene_name": kinase_gene_name
        }
        t_df = pd.DataFrame([t_dict])
        df2 = pd.concat([df2, t_df], ignore_index=True)


df2
```

```python
categories = ["molecular_function", "biological_process", "cellular_component"] #categories to be included

#list to include values in each categories

hpga_df = pd.read_csv("human_protein_go_annotation.csv")

for protein_accession in df['uniprotkb_canonical_ac']: #for each protein in the original data set
    biological_list = []
    molecular_list = []
    cellular_list = []
    protein_df = hpga_df[hpga_df['uniprotkb_canonical_ac'] == protein_accession]
    for category in categories: #extracting info based on categories into the corresponsing list
        go_terms = protein_df.loc[protein_df['go_term_category'] == category, ['go_term_label', 'go_term_id']]
        if category == "molecular_function":
            for index, row in go_terms.iterrows():
                molecular_list.append(f"{row['go_term_label']} ({row['go_term_id']})")
        elif category == "biological_process":
            for index, row in go_terms.iterrows():
                biological_list.append(f"{row['go_term_label']} ({row['go_term_id']})")
        elif category == "cellular_component":
            for index, row in go_terms.iterrows():
                cellular_list.append(f"{row['go_term_label']} ({row['go_term_id']})")

    biological_process = ', '.join(map(str, biological_list)) #converting to string and adding ','
    molecular_function = ', '.join(map(str, molecular_list))
    cellular_component = ', '.join(map(str, cellular_list))


    df2.loc[df2.uniprotkb_canonical_ac == protein_accession,["molecular_function"]] = molecular_function #adding info to the big dataset df2
    df2.loc[df2.uniprotkb_canonical_ac == protein_accession,['biological_process']] = biological_process
    df2.loc[df2.uniprotkb_canonical_ac == protein_accession,['cellular_component']] = cellular_component


df2
```
